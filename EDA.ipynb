# Are there some continents that have grown faster than others and in which ways??

```python
# First lets group the continents. 
df.groupby('Continent').mean()
```

```python
# Focus on only the continent Oceania 
df[df['Continent'].str.contains('Oceania')]  #They seem like islands, small continents
```

```python
# sort continents in an Descending order to confirm the small continents
df2 = df.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)
df2
```

```python
# Plot the diagram
df2.plot()
```

## That looks different so let's try re order it to make sense


```python
df.columns
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending = False)
```

```python
# Transpose
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
